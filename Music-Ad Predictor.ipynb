```python
import matplotlib
import pyaudio
import numpy
import scipy
import scipy.io.wavfile as wav
from os import listdir
from os.path import isfile, join
from matplotlib import figure
from __future__ import division
import wave as wv
import beat_detection as bt
import csv


advertfolder = 'Advert'
musicfolder = 'Music'

radio_data_file =  open('radio_data.csv', 'wb')
radio_data_csv = csv.writer(radio_data_file, delimiter=',')


%pylab inline

adfiles = [ f for f in listdir(advertfolder) if isfile(join(advertfolder,f)) ]
musicfiles = [ f for f in listdir(musicfolder) if isfile(join(musicfolder,f)) ]

data = numpy.zeros((len(adfiles) + len(musicfiles), 4))
result = numpy.zeros(len(adfiles) + len(musicfiles))
counter = 0

def processFile(wavefile, filetype):
    (rate,sample) = wav.read(wavefile)
    return (rate,sample)

def get_wav_info(wav_file):
    wav = wv.open(wav_file, 'r')
    frames = wav.readframes(-1)
    sound_info = pylab.fromstring(frames, 'Int16')
    frame_rate = wav.getframerate()
    wav.close()
    return frame_rate,sound_info
```

```python
def setup_graph(title='', x_label='', y_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0], fig_size[1])
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
```

```python
def plotwave(sample_rate, input_signal):
    length = float(len(input_signal))
    time_array = numpy.arange(0, length/sample_rate, 1.0/sample_rate)
    setup_graph(title='1', x_label='time (in seconds)', y_label='amplitude', fig_size=(14,2))
    _ = plot(time_array[0:500000], input_signal[0:500000])
    
```

```python
def compute_fft(sample_rate,input_signal):
    fft_out = numpy.fft.rfft(input_signal)
    fft_mag = [sqrt(i.real**2 + i.imag**2)/len(fft_out) for i in fft_out]
    num_samples = len(input_signal)
    rfreqs = [(i*1.0/num_samples)*sample_rate for i in range(num_samples//2+1)]   
    return (rfreqs, fft_mag)

def plot_fft(rfreqs, fft_mag):
    setup_graph(title='F', x_label='FFT Bins', y_label='magnitude', fig_size=(14,2))
    _ = plot(rfreqs[0:40000], fft_mag[0:40000])
```
